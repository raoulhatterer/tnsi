### 2. Diffusion
Le code ci-dessous va créer une grille comportant un motif :

```python
from ipythonblocks import BlockGrid

n = 10

def initgrid():
    
    grid = BlockGrid(n, n, block_size=15, fill = (180,180,180))
    grid[1,3].set_colors(0,0,0)
    grid[2,4].set_colors(0,0,0)
    grid[2,5].set_colors(0,0,0)
    grid[1,6].set_colors(0,0,0)
    grid[2,2].set_colors(0,0,0)
    grid[2,7].set_colors(0,0,0)
    grid[3,7].set_colors(0,0,0)
    grid[3,2].set_colors(0,0,0)
    grid[4,3].set_colors(0,0,0)
    grid[4,6].set_colors(0,0,0)
    grid[5,2].set_colors(0,0,0)
    grid[5,7].set_colors(0,0,0)
    grid[6,1].set_colors(0,0,0)
    grid[6,8].set_colors(0,0,0)
    grid[7,2].set_colors(0,0,0)
    grid[7,3].set_colors(0,0,0)
    grid[6,4].set_colors(0,0,0)
    grid[6,5].set_colors(0,0,0)
    grid[7,6].set_colors(0,0,0)
    grid[7,7].set_colors(0,0,0)   
    
    return grid

grid = initgrid()
grid.show()
```

Écrire une fonction `propager(grid,i,j,color)` qui va modifier **la composante verte** de chaque cellule voisine de couleur identique à la cellule `[i,j]`. 

On travaillera uniquement avec la composante verte pour la détection/modification des cellules.

```python
def propager(grid, i, j, color):

    if grid[i,j].green == 0:
        return None # 
    
    grid[i,j].green = color

    # l'élément en haut fait partie de la composante
    if ((i-1) >= 0 and grid[i-1,j].green == 180):
        propager(grid, i-1, j, color)

    # l'élément en bas fait partie de la composante
    if ((i+1) < n and grid[i+1,j].green == 180):
        propager(grid, i+1, j, color)

    # l'élément à gauche fait partie de la composante
    if ((j-1) >= 0 and grid[i,j-1].green == 180):
        propager(grid, i, j-1, color)

    # l'élément à droite fait partie de la composante
    if ((j+1) < n and grid[i,j+1].green == 180): # 
        propager(grid, i, j+1, color)

   
grid = initgrid()
propager(grid,3,3,0) # on colorie la cellule [3,3] en mettant sa composante verte à 0
grid.show()


```
